## TEST CASE 2: Validate Image Input and Output


This test case validates that the software has the ability to:
1. Correctly load a `.tif` image.
2. Extract important metadata from the image, such as width, height, and color mode.
3. Export this metadata into a CSV file to simulate structured data output.

This ensures that the software can process high-resolution images and output the expected structured data, which is crucial for ensuring data integrity and further analysis.


```python

# Load the image
image_path = 'AS_09125_050118150001_A03f05d0.tif'
img = Image.open(image_path)

# Extract data (e.g., dimensions, mode)
data = {
    'width': img.width,
    'height': img.height,
    'mode': img.mode
}

data
```

```python
# output to CSV
csv_file = 'output.csv'
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Attribute', 'Value'])
    for key, value in data.items():
        writer.writerow([key, value])

print(f"CSV {csv_file} generated successfully.")
```

## TEST CASE 3: Invalid Image Format Handling

This test case, focuses on testing the software's ability to handle unsupported or invalid image formats. The goal is to verify that the software correctly identifies and rejects images that do not conform to the expected `.tif` format, such as `.pdf`, `.jpeg`, or corrupted files.


```python

try:
    img = Image.open('test_file.pdf')  # Invalid image format
    img.verify()  # Check if image is valid
    print("Image is valid.")
except (UnidentifiedImageError, IOError):
    print("Test passed: Invalid image format rejected.")
```
